```python
# for two populations from standard normal distribution, do a t-test for M=10000 times
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

n1, n2 = 100, 100
M = 50000
mu1, mu2 = 0, 0
sigma1, sigma2 = 1, 1
alpha = 0.05
s1, s2 = np.sqrt(sigma1), np.sqrt(sigma2)
X1 = np.random.normal(mu1, s1, (n1, M))
X2 = np.random.normal(mu2, s2, (n2, M))
results = ttest_ind(X1, X2, equal_var = True)
plt.hist(results.pvalue, edgecolor = 'y')
plt.show()
plt.hist(results.statistic, edgecolor = 'y', bins = 100)
plt.show()
power = np.mean(results.pvalue < alpha)
print(power)
# print(np.mean(p < 0.05))
# print(results)
# print(results.statistic)
# print(results.pvalue)
# print(results.df)
```
