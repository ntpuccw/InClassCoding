```python
# generate N samples from a normal distribution with mean 0 and variance 1
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

N = 10000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N) # generate N samples from normal distribution with mean 0 and variance 1
plt.figure(figsize = (6, 3))
plt.hist(x, edgecolor = 'b', bins = 10, density = True)
plt.xlabel('x')
plt.show()

# plot a pdf of the normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
x = np.linspace(-5, 5, 100)
plt.plot(x, norm.pdf(x))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Number of samples
N = 100000

# Generate N samples from a normal distribution with mean 0 and variance 1
samples = np.random.normal(0, 1, N)

# Plot the histogram of the samples
plt.figure(figsize=(6, 3))
plt.hist(samples, bins=30, edgecolor='b', density=True, alpha=0.3, color='b')

# Plot the PDF of the normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'r', linewidth=2, alpha=0.6)

title = "Histogram and PDF of Normal Distribution"
plt.title(title)
plt.xlabel('$\mu$')
plt.ylabel('Density')
plt.legend([ 'PDF', 'Histogram'])
plt.grid(True)
plt.show()
```

```python
# make a 2x2 grid of subplots
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
fig, axs = plt.subplots(2, 2, figsize=(8, 4))

# generate N samples from a normal distribution with mean 0 and variance 1
N = 100000
mu = 0
sigma = 1
# x = norm.rvs(loc = mu, scale = sigma, size = N)
# generate N samples from chi-square distribution with df = 4
x = np.random.chisquare(df = 4, size = N)

# plot a histogram of the samples
axs[0, 0].hist(x, edgecolor = 'b', bins = 10, density = True)
axs[0, 0].set_xlabel('x')

# plot a pdf of the normal distribution
x = np.linspace(-5, 5, 100)
axs[0, 1].plot(x, norm.pdf(x))
axs[0, 1].set_xlabel('$\mu$')

# plot a boxplot of the samples
axs[1, 0].boxplot(x)
axs[1, 0].set_xlabel('x')
# plot an empirical CDF of the samples
axs[1, 1].plot(np.sort(x), np.linspace(0, 1, 100))
axs[1, 1].set_xlabel('x')

plt.show()
```

```python
# 畫一張圖含有兩個子圖，左邊是一個直方圖，右邊是一個箱形圖
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

N = 100000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N)
axs[0].hist(x, edgecolor = 'b', bins = 10, density = True)
axs[0].set_xlabel('x')

axs[1].boxplot(x)
axs[1].set_xlabel('x')
plt.show()
```

```python
# 畫一張圖含有 10 個卡方分配的PDF，自由度從 1 到 10
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
x = np.linspace(0, 40, 1000)
for df in range(4, 20):
    plt.plot(x, chi2.pdf(x, df), label = f'df = {df}')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.show()

# 畫一張圖含有 10 個卡方分配的PDF，自由度從 1 到 10，但是利用 broadcasting 的方式
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
x = np.linspace(0, 60, 1000)
df = np.arange(4, 40, 2)
# Y = chi2.pdf(x[:, np.newaxis], df)
Y = chi2.pdf(x.reshape(-1, 1), df)
label = [f'df = {i}' for i in df]
plt.plot(x, Y, color = 'b', alpha = 0.3, label = label)
# plt.plot(x, chi2.pdf(x[:, np.newaxis], df))
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.show()
```
