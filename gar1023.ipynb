```python
# randomly select 4 numbers from a=[2, 4, 9, 12] for 1000 times

import numpy as np
import matplotlib.pyplot as plt

N =1000000 # sample times
a = np.array([2, 4, 9, 12])
X = np.random.choice(a = a, size=(N, 4))
y = X.mean(axis=1)
# count the number of each value in y
values, counts = np.unique(y, return_counts=True)
# draw bar chart
plt.style.use('ggplot')
fig = plt.figure(figsize=[6, 4])
plt.bar(values, counts / N, width=0.1)
plt.show()
y_count = dict(zip(values, counts))

```

```python
# generate 100 samples from a normal distribution with mean 0 and variance 1
# import scipy.stats as stats
from scipy.stats import norm 
from scipy import stats

N = 100
M = 1000
mu1 = 0
mu2 = 0.2
sigma = 1
P_Val = np.zeros((M, 1))
for i in range(M):
    X1 = norm.rvs(loc = mu1, scale = sigma, size = N)
    X2 = norm.rvs(loc = mu2, scale = sigma, size = N)
# do a t-test and print the p-value
    results = stats.ttest_ind(X1, X2)
    P_Val[i] = results[1]
# print the p-value
# print(results[1])
plt.hist(P_Val, bins=50)
plt.show()
```
