****<font color=skyblue>繪製常態分配的 PDF 函數</font>****

$$f(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^{2}}{2}}$$

1. 繪製常態分配的 PDF 函數
1. 繪製常態分配的 CDF 函數
<hr>

```python
# draw a pdf of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

# x = np.arange(-10, 10, 0.01) 
x = np.linspace(-5, 5, 1000) # generate 1000 equally spaced numbers between -10 and 10
y = stats.norm.pdf(x, loc = 0, scale = 1) # loc is the mean, scale is the standard deviation

plt.figure(figsize = (6, 3))
plt.style.use('ggplot')
# plt.style.use('fivethirtyeight')
# plt.style.use('dark_background')

plt.plot(x, y, linewidth = 3, color = 'r', \
         label = 'Normal Distribution', alpha = 0.5, linestyle = '-')

plt.legend()
plt.grid(True)
plt.title('$f(x) = \\frac{1}{\\sigma\\sqrt{2\\pi}}e^{-\\frac{(x-\\mu)^2}{2\\sigma^2}}$')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

```python
# generate N samples from normal distribution with mean 0 and variance 1
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
N = 1000000
mu = 0
sigma = 1
x = norm.rvs(loc = mu, scale = sigma, size = N)
plt.figure(figsize = (6, 3))
plt.hist(x, edgecolor = 'b', bins = 200, density = True)
plt.xlabel('x')
plt.show()
```
