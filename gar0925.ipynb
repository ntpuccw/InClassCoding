```python
# generate n samples from chi-square distribution with k degrees of freedom

import numpy as np
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt

n = 100 # sample size
k = 2 # degrees of freedom
N = 100000 # number of samples

# x_bar = np.zeros(N)
# for i in range(N):
#     # x = chi2.rvs(k, size=n)
#     x = norm.rvs(size=n)
#     x_bar[i] = np.mean(x)

# do the same thing without looping
X = chi2.rvs(k, size=(N, n))
x_bar = np.mean(X, axis=1)

fig = plt.figure(figsize = (6, 3))
plt.hist(x_bar, bins = 50, density = True, color = 'orange', alpha = 0.7)
plt.show()
```

```python
# draw the pdf of T distribution
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# degrees of freedom from 0.1:0.1:2.0 and 1:3:30
k = np.arange(0.1, 2, 0.1)
k = np.append(k, np.arange(1, 30, 3))
x = np.linspace(-6, 6, 1000)
fig = plt.figure(figsize = (6, 3))
# draw the pdf of standard normal distribution
z = norm.pdf(x, 0, 1)
plt.plot(x, z, linewidth = 3, color = 'r', label = 'Z distribution')

# for i in range(len(k)):
#     y = t.pdf(x, k[i])
#     # plt.plot(x, y, color = 'b', alpha = 0.5, label = 'k = ' + str(k[i]))
#     plt.plot(x, y, color = 'b', alpha = 0.5)

# do the same thing in a broadcast way
x = x.reshape(-1, 1)
Y = t.pdf(x, k) # Y = t.pdf(x, k.reshape(1, -1))
# Y = t.pdf(x.reshape(-1, 1), k.reshape(1, -1))
# Y = t.pdf(x[:, np.newaxis], k)
plt.plot(x, Y, color = 'b', alpha = 0.5)

plt.title('The t probability distribution function')
plt.legend()
plt.grid(True)
plt.show()
```
