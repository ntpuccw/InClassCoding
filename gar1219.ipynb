```python
# generate N samples of a random variable X with beta(a,b) distribution

import numpy as np
import matplotlib.pyplot as plt
# import scipy.stats as stats
from scipy.stats import beta

# parameters
a1, b1, a2, b2 = 9, 2, 8, 2
N = 10000

# generate samples
# X = np.random.beta(a1, b1, N)
# Y = np.random.beta(a2, b2, N)
X = beta.rvs(a1, b1, size=N)
Y = beta.rvs(a2, b2, size=N)
Z = X * Y
# use MLE to estimate the parameters of the beta distribution form Z
a, b, loc, scale = beta.fit(Z, method='MM' floc=0, fscale=1)
print('a = %f, b = %f' % (a, b))

z = np.linspace(0, 1, 1000)
plt.plot(z, beta.pdf(z, a1, b1), 'r-', lw=2, alpha=0.6, label='beta pdf (a=9, b=2)') 
plt.plot(z, beta.pdf(z, a2, b2), 'b-', lw=2, alpha=0.6, label='beta pdf (a=8, b=2)')
plt.hist(Z, bins=50, density=True, alpha=0.6, color='g')
plt.plot(z, beta.pdf(z, a, b), 'k-', lw=2, alpha=0.6, label='beta pdf (a=%f, b=%f)' % (a, b))
plt.show()

```

```python
from scipy.stats import beta

def unit_serial(a1, b1, a2, b2):
    # generate samples
    N = 50000
    X = beta.rvs(a1, b1, size=N)
    Y = beta.rvs(a2, b2, size=N)
    Z = X * Y
    # use MLE to estimate the parameters of the beta distribution form Z
    a, b, loc, scale = beta.fit(Z, method='MM', floc=0, fscale=1)
    return a, b

a1, b1, a2, b2 = 9, 2, 8, 2
print(unit_serial(a1, b1, a2, b2))
```
