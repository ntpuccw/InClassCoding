Find the minimum value of the function $f(x) = x^2 + 3x + 5$

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define the polynomial function
# def f(x):
#     return x**2 + 3*x + 5
def f(x):
    coef = [1, -8, 16, -2, 8]
    return np.polyval(coef, x)
    # return x**4 - 8*x**3 + 16*x**2 - 2*x + 8

# draw the graph
x = np.linspace(-1, 5, 1000)
y = f(x)
plt.plot(x, y)
plt.xlabel('X'), plt.grid(True)
plt.show()

# Initial guess for the minimum
x0 = 1
# Use scipy.optimize.minimize to find the minimum
result = minimize(f, x0, method='BFGS', tol=1e-8)

# Print the result
print(f"Minimum value of the function: {result.fun}")
print(f"Value of x at the minimum: {result.x[0]}")
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the function
def f(vars):
    x, y = vars
    return x**2 + y**2 + 3*x + 5*y

# Define the constraint y > 0
def constraint(vars):
    return vars[1]  # y must be greater than 0

# Draw the contour plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2 + 3*X + 5*Y
plt.contour(X, Y, Z, 100)
plt.xlabel('X'), plt.ylabel('Y'), plt.grid(True)
plt.show()


# Initial guess for the minimum
x0 = [0, 1]

# Define the constraint dictionary
cons = {'type': 'ineq', 'fun': constraint}

# Use scipy.optimize.minimize to find the minimum
result = minimize(f, x0, constraints=cons)

# Print the result
print(f"Minimum value of the function: {result.fun}")
print(f"Values of x and y at the minimum: {result.x}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of N
N_values = [10, 30, 50, 100, 500, 1000]

# Mean and standard deviation of the normal distribution
mu, sigma = 0, 1

# Compute the average for each N
averages = []
for N in N_values:
    samples = np.random.normal(mu, sigma, N)
    average = np.mean(samples)
    averages.append(average)

# Plot the average vs N
fig, ax = plt.subplots(figsize=(6, 3))
plt.figure(figsize=(10, 6))
ax.plot(averages, marker='o', linestyle='-', color='b')
ax.set_xticks(np.arange(len(N_values)))
ax.set_xticklabels(N_values)
ax.set_xlabel('N')
ax.set_ylabel('Average')
ax.grid(True)
plt.xlabel('N')
plt.ylabel('Average')
plt.title('Average of Samples vs N')
plt.grid(True)
ax.set_ylim(-1, 1)
plt.show()
```
