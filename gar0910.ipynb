```python
# plot a graph of a function
# y = f(x)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100) # 100 points from -10 to 10
y = x**2 # y = f(x) = x^2
fig = plt.figure(figsize=(4, 3)) # 4 inches by 3 inches
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = x^2$')
plt.grid()
plt.show()
```

```python
# plot a normal distribution curve
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np


mu = 0 # mean
sigma = 1 # standard deviation
x = np.linspace(-5, 5, 100)
# use pdf command from scipy.stats
y = norm.cdf(x, mu, sigma)
# y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2)
fig = plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution Curve')
plt.grid()
plt.show()
```

```python
# plot 5 normal distribution curves with different means and standard deviations=1
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats 

sigma = 1 # standard deviation
x = np.linspace(-5, 10, 100)

fig = plt.figure(figsize=(10, 3))

for i in range(5):
    mu = i # mean
    y = stats.norm.pdf(x, mu, sigma)
    # use latex in label
    # y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2) 
    plt.plot(x, y, label=f'$\mu$={mu}, $\sigma$={sigma}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution Curves with Different Means')
plt.legend()
plt.grid()
plt.show()
```
