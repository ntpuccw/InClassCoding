```python
# plot a graph of a function
# y = f(x)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100) # 100 points from -10 to 10
y = x**2 #
fig = plt.figure(figsize=(4, 3)) # 4 inches by 3 inches
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = x^2$')
plt.grid()
plt.show()
```

$$f(x) = \int x^{2} dx$$
