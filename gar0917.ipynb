```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats 
```

```python
# plot 5 normal distribution curves with different means and standard deviations=1


sigma = 1 # standard deviation
x_left, x_right = -5, 10
n = 1000
x = np.linspace(x_left, x_right, n)
# -----------------------------------------------
fig = plt.figure(figsize=(10, 3))

for i in range(5):
    mu = i # mean
    y = stats.norm.pdf(x, mu, sigma)
    # use latex in label
    # y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu)/sigma)**2) 
    plt.plot(x, y, label=f'$\mu$={mu}, $\sigma$={sigma}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Distribution Curves with Different Means')
plt.legend()
plt.grid()
plt.show()
```

```python
# 
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11, 4))
xlim = [-2.5, 8.5]
sigma = 1
mu = np.arange(1, 6)

x = np.linspace(xlim[0], xlim[1], 1000)
x_c = x.reshape(-1, 1) # column vector
Y = stats.norm.pdf(x_c, mu, sigma) # broadcasting
label = ["$\mu$={}".format(i) for i in mu] # list comprehension
ax[0].plot(x, Y, label=label)
ax[0].legend()
ax[0].grid()
ax[0].set_title("Normal Distribution with $\sigma=${}".format(sigma))

xlim = [-15, 15] #  sigma 
mu = 0
s = np.arange(1, 6)
x = np.linspace(xlim[0], xlim[1], 1000)
Y = stats.norm.pdf(x.reshape(-1, 1), mu, s)
label = ["$\sigma$={}".format(i) for i in s]
ax[1].plot(x, Y, label=label)
ax[1].legend()
ax[1].grid()
ax[1].set_title("Normal Distribution with $\mu=${}".format(mu))
plt.show()
```

```python
# plot chi-square distribution with different degrees of freedom
fig, ax = plt.subplots(figsize=(8, 3))
x_left, x_right = 0, 50
df = np.arange(4, 30, 1) # degrees of freedom
x = np.linspace(x_left, x_right, 1000)

for i in df:
    y = stats.chi2.pdf(x, df=i)
    ax.plot(x, y, color="blue", label="$df$={}".format(i))
# ax.legend()
ax.grid()
ax.set_title("Chi-Square Distribution with Different Degrees of Freedom")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
# plot chi-square distribution with different degrees of freedom
fig, ax = plt.subplots(figsize=(6, 3))
x_left, x_right = -5, 5
df = np.arange(1, 10, 1) # degrees of freedom
x = np.linspace(x_left, x_right, 1000)
# use boradcasting to plot multiple curves
y = stats.t.pdf(x.reshape(-1, 1), df=df)
ax.plot(x, y, color="blue", label="$df$={}".format(df))
ax.grid()
ax.set_title("T Distribution with Different Degrees of Freedom")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
# gernerate random numbers from normal distribution
n = 100
bins = 10
x = stats.norm.rvs(loc=0, scale=1, size=n)

fig = plt.figure(figsize=(6, 3))
plt.hist(x, bins=bins, density=True, alpha=0.4, color='b', edgecolor='k')
plt.title('Sample size = {}'.format(n))
# plot the PDF
xx = np.linspace(-5, 5, 1000)
y = stats.norm.pdf(xx, loc=0, scale=1)
plt.plot(xx, y, color='r')
plt.show()
```

```python
# gernerate random numbers from chi square distribution and plot boxplot
n = 1000
df = 10
x = stats.chi2.rvs(df=df, size=n)
fig = plt.figure(figsize=(3, 3))
plt.boxplot(x)
plt.title('Sample size = {}'.format(n))
plt.show()
```
