```python

# load the excel file in the directory data/ and print the first 5 rows
import pandas as pd
import numpy as np

df = pd.read_excel('data/city_quality.xlsx')
df.head() # print the first 5 rows
df.describe() # print the summary statistics
# df.info() # print the data types of each column
# df.columns # print the column names
# df.shape # print the number of rows and columns
```

```python
#  draw a boxplot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
# sns.boxplot(data=df)
# draw in a horizontal way
sns.boxplot(data=df, orient='h')
plt.title('City Quality')
plt.xlabel('City')
plt.ylabel('Quality')
plt.show()
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # create a StandardScaler object
scaler.fit(df) # fit the data
scaled_df = scaler.transform(df) #
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)

# draw a boxplot
plt.figure(figsize=(6, 4))
sns.boxplot(data=scaled_df, orient='h')
plt.title('City Quality')
plt.xlabel('City')
plt.ylabel('Quality')
plt.show()
```
